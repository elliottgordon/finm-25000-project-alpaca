```python
from alpaca.trading.client import TradingClient
import os
import sys

sys.path.append(os.path.abspath('..'))
from Alpaca_API import ALPACA_KEY
from Alpaca_API import ALPACA_SECRET

# paper=True enables paper trading
trading_client = TradingClient(ALPACA_KEY, ALPACA_SECRET, paper=True)

account = trading_client.get_account()
print(f"Account ID: {account.id}")
print(f"Account Status: {account.status}")
print(f"Account Cash: {account.cash}")
print(f"Account Equity: {account.equity}")
print(f"Account Buying Power: {account.buying_power}")
```

```python
from alpaca.trading.requests import MarketOrderRequest
from alpaca.trading.enums import OrderSide, TimeInForce

# Preparing orders
market_order_data = MarketOrderRequest(
    symbol="SPY",
    qty = 1,
    side = OrderSide.BUY,
    time_in_force = TimeInForce.DAY
)

# Market order
market_order = trading_client.submit_order(order_data=market_order_data)
print(f"Market Order ID: {market_order.id}")
print(f"Market Order Status: {market_order.status}")
print(f"Market Order Filled Avg Price: {market_order.filled_avg_price}")
print(f"Market Order Filled Qty: {market_order.filled_qty}")
```

```python
# Cancel the order you just placed using its ID
try:
    cancel_response = trading_client.cancel_order_by_id(market_order.id)
    print(f"Order {market_order.id} has been cancelled")
    print(f"Cancel status: {cancel_response}")
except Exception as e:
    print(f"Error cancelling order: {e}")
```
